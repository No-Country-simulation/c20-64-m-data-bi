```python
import pandas as pd
import numpy as nm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_excel('/Municipios.xlsx')
df2 = pd.read_excel('/PP_Chiapas.xlsx')
df3 = pd.read_excel('/Prom_Temperatura_Chiapas.xlsx')
df4 = pd.read_excel('/Suelos_Chiapas.xlsx')
df5 = pd.read_excel('/cierre_agricola_chiapas.xlsx')

```

```python
df1.head()
```

```python
df2.head()
```

```python
df2['PP_Anual']= df2[['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']].mean(axis=1).round(2)
```

```python
df2.head()
```

```python
# Selección de dos columnas del DataFrame
Promedio_Precip_Anual= df2[['Año', 'PP_Anual']]
print(Promedio_Precip_Anual)
```

```python
df3.head()
#Temperatura
```

```python
# Selección de dos columnas del DataFrame
Promedio_temperatura_anual= df3[['Año', 'Anual']]
print(Promedio_temperatura_anual)
```

```python
df4.head()
```

```python
df5 = df5.rename(columns={'Anio': 'Año'})
df5.head()
```

```python
total_produccion_anual = df5.groupby('Año')['Volumenproduccion'].sum().reset_index()
print(total_produccion_anual)
```

```python
Promedio_Producción_Anual = df5.groupby('Año')['Volumenproduccion'].mean().reset_index()
print(Promedio_Producción_Anual)
```

```python
años = list(range(2003, 2024))
plt.figure(figsize=(12, 8))
plt.plot(años, Promedio_Producción_Anual, label='Producción Agrícola', marker='o')
plt.plot(años, Promedio_temperatura_anual, label='Temperatura', marker='o')
plt.plot(años, Promedio_Precip_Anual, label='Precipitación', marker='o')

plt.xlabel('Año')
plt.ylabel('Valor')
plt.title('Evolución Anual de Producción Agrícola, Temperatura y Precipitación')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Combinación de las tablas
df_combinadas = pd.merge(Promedio_Precip_Anual, Promedio_temperatura_anual, on='Año')
df_combinadas = pd.merge(df_combinadas,Promedio_Producción_Anual , on='Año')
print(df_combinadas)
```

Resultado: El resultado es una nueva tabla (DataFrame) que muestra los coeficientes de correlación de Pearson entre cada par de columnas. Los valores oscilan entre -1 y 1:

1: correlación positiva perfecta.
0: sin correlación.
-1: correlación negativa perfecta.

```python
# Calcular la matriz de correlación
correlation_matrix = df_combinadas.corr()
print(correlation_matrix)
```

```python
#grafico de dispersión
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_combinadas, x='PP_Anual', y='Anual', hue='Volumenproduccion')
sns.regplot(data=df_combinadas, x='PP_Anual', y='Anual', scatter=False, color='black', line_kws={'label': 'Línea de Tendencia'})
for i in range(len(df_combinadas)):
    plt.text(df_combinadas['PP_Anual'][i], df_combinadas['Anual'][i], str(df_combinadas['Año'][i]), fontsize=10, ha='right')
plt.xlabel('Precipitación(mm)')
plt.ylabel('Temperatura (°C)')
plt.title('Relación entre Precipitación, Temperatura y Producción Agrícola')
plt.legend(title='Producción Agrícola')
plt.grid(True)
plt.show()
```

```python
# Selecciono solo las dos columnas
df_seleccionado = df_combinadas[['Volumenproduccion', 'Anual']]

# Calcula la correlación
correlacion = df_seleccionado.corr()

print(correlacion)
```

Analisis: segun la interpretación del analisis de correlación de pearson, existe una correlacion negativa moderada o inversa, hay una tendencia general en los datos, es decir que a medida que la temperatura aumenta la producción tiende a disminuir.

```python
#grafico de dispersión Producción vs Temperatura
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_combinadas, x='Anual', y='Volumenproduccion', hue='Volumenproduccion')
sns.regplot(data=df_combinadas, x='Anual', y='Volumenproduccion', scatter=False, color='black', line_kws={'label': 'Línea de Tendencia'})
plt.ylabel('Producción Agrícola')
plt.title('Relación entre Temperatura y Producción Agrícola')
plt.legend(title='Producción Agrícola')
plt.grid(True)
plt.show()
```

```python
# Selecciono solo las dos columnas-Precipitación
df_seleccionado = df_combinadas[['Volumenproduccion', 'PP_Anual']]

# Calcula la correlación
correlacion = df_seleccionado.corr()

print(correlacion)
```

Hay una relación positiva entre las dos variables. Esto significa que, en general, a medida que una variable aumenta, la otra también tiende a aumentar. Sin embargo, esta relación es débil, aunque existe una tendencia observable , hay mucha variabilidad en los datos, lo que significa que no se puede predecir con precisión el comportamiento de una variable basándose en la otra.
Ambas tienden a aumentar juntas, pero la relación no es lo suficientemente fuerte como para hacer predicciones fiables.

```python
#grafico de dispersión Producción vs precipitación
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_combinadas, x='PP_Anual', y='Volumenproduccion', hue='Volumenproduccion')
sns.regplot(data=df_combinadas, x='PP_Anual', y='Volumenproduccion', scatter=False, color='black', line_kws={'label': 'Línea de Tendencia'})
plt.ylabel('Producción Agrícola')
plt.title('Relación entre Precipitación y Producción Agrícola')
plt.legend(title='Producción Agrícola')
plt.grid(True)
plt.show()
```
