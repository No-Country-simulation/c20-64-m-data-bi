```python
# Importar pandas
import pandas as pd

# Importar Matplotlib y Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
cierre_agricola = pd.read_csv('C:/DAVINIA/NoCountry/Proyecto/cierre_agricola_chiapas.csv', encoding='utf-8-sig')
cierre_agricola.head()
```

Corrección de datos

```python
# Reemplazar comas por puntos
cierre_agricola['Sembrada'] = cierre_agricola['Sembrada'].str.replace(',', '.')
cierre_agricola['Cosechada'] = cierre_agricola['Cosechada'].str.replace(',', '.')
cierre_agricola['Volumenproduccion'] = cierre_agricola['Volumenproduccion'].str.replace(',', '.')
cierre_agricola['Valorproduccion'] = cierre_agricola['Valorproduccion'].str.replace(',', '.')

# Convertir datos a numéricos
cierre_agricola['Sembrada'] = pd.to_numeric(cierre_agricola['Sembrada'], errors='coerce')
cierre_agricola['Cosechada'] = pd.to_numeric(cierre_agricola['Cosechada'], errors='coerce')
cierre_agricola['Volumenproduccion'] = pd.to_numeric(cierre_agricola['Volumenproduccion'], errors='coerce')
cierre_agricola['Valorproduccion'] = pd.to_numeric(cierre_agricola['Valorproduccion'], errors='coerce') / 1000
```

Filtración de los datos por cultivo

```python
# Filtrar los datos para el cultivo
maiz = cierre_agricola[cierre_agricola['Nomcultivo'] == 'Maíz grano']
cana_azucar = cierre_agricola[cierre_agricola['Nomcultivo'] == 'Caña de azúcar']
cafe_cereza = cierre_agricola[cierre_agricola['Nomcultivo'] == 'Café cereza']
platano = cierre_agricola[cierre_agricola['Nomcultivo'] == 'Plátano']
mango = cierre_agricola[cierre_agricola['Nomcultivo'] == 'Mango']
frijol = cierre_agricola[cierre_agricola['Nomcultivo'] == 'Frijol']
papaya = cierre_agricola[cierre_agricola['Nomcultivo'] == 'Papaya']
semilla_soya = cierre_agricola[cierre_agricola['Nomcultivo'] == 'Semilla de soya']
tomate_rojo = cierre_agricola[cierre_agricola['Nomcultivo'] == 'Tomate rojo (jitomate)']
cacao = cierre_agricola[cierre_agricola['Nomcultivo'] == 'Cacao']
aguacate = cierre_agricola[cierre_agricola['Nomcultivo'] == 'Aguacate']
papa = cierre_agricola[cierre_agricola['Nomcultivo'] == 'Papa']
```

Agrupación según area sembrada, cosechada, volumen y valor total de producción

```python
# Agrupar por año y calcular las áreas totales sembradas, cosechadas y el volumen total de producción.
maiz_grouped = maiz.groupby('Anio', as_index=False)[['Sembrada', 'Cosechada', 'Volumenproduccion', 'Valorproduccion']].sum()
cana_azucar_grouped = cana_azucar.groupby('Anio', as_index=False)[['Sembrada', 'Cosechada', 'Volumenproduccion', 'Valorproduccion']].sum()
cafe_cereza_grouped = cafe_cereza.groupby('Anio', as_index=False)[['Sembrada', 'Cosechada', 'Volumenproduccion', 'Valorproduccion']].sum()
platano_grouped = platano.groupby('Anio', as_index=False)[['Sembrada', 'Cosechada', 'Volumenproduccion', 'Valorproduccion']].sum()
mango_grouped = mango.groupby('Anio', as_index=False)[['Sembrada', 'Cosechada', 'Volumenproduccion', 'Valorproduccion']].sum()
frijol_grouped = frijol.groupby('Anio', as_index=False)[['Sembrada', 'Cosechada', 'Volumenproduccion', 'Valorproduccion']].sum()
papaya_grouped = papaya.groupby('Anio', as_index=False)[['Sembrada', 'Cosechada', 'Volumenproduccion', 'Valorproduccion']].sum()
semilla_soya_grouped = semilla_soya.groupby('Anio', as_index=False)[['Sembrada', 'Cosechada', 'Volumenproduccion', 'Valorproduccion']].sum()
tomate_rojo_grouped = tomate_rojo.groupby('Anio', as_index=False)[['Sembrada', 'Cosechada', 'Volumenproduccion', 'Valorproduccion']].sum()
cacao_grouped = cacao.groupby('Anio', as_index=False)[['Sembrada', 'Cosechada', 'Volumenproduccion', 'Valorproduccion']].sum()
aguacate_grouped = aguacate.groupby('Anio', as_index=False)[['Sembrada', 'Cosechada', 'Volumenproduccion', 'Valorproduccion']].sum()
papa_grouped = papa.groupby('Anio', as_index=False)[['Sembrada', 'Cosechada', 'Volumenproduccion', 'Valorproduccion']].sum()
```

## Graficos según cultivo

### Grafico Maiz

```python
# Configurar el estilo de Seaborn
sns.set_style('whitegrid')

# Crear el gráfico
plt.figure(figsize=(12, 6))
sns.lineplot(data=maiz_grouped, x='Anio', y='Sembrada', marker='o', label='Superficie Sembrada')
sns.lineplot(data=maiz_grouped, x='Anio', y='Cosechada', marker='o', label='Superficie Cosechada')
sns.lineplot(data=maiz_grouped, x='Anio', y='Volumenproduccion', marker='o', label='Volumen de Producción', linestyle='--')
sns.lineplot(data=maiz_grouped, x='Anio', y='Valorproduccion', marker='o', label='Valor de Producción', linestyle='--')

# Personalizar el gráfico
plt.title('Volumen y Valor de Producción del Maíz Grano en Chiapas')
plt.xlabel('Año')
plt.ylabel('Superficie (hectáreas) / Volumen (toneladas) / Valor (miles de pesos)')
plt.legend()

# Asegurarse de que los años en el eje x aparezcan como enteros
plt.xticks(ticks=range(int(maiz_grouped['Anio'].min()), int(maiz_grouped['Anio'].max()) + 2, 2))

# Mostrar el gráfico
plt.show()
```
