## TENDENCIAS A LO LARGO DEL TIEMPO

```python
# Paso 1: Agrupar los datos por año y cultivo, sumando el volumen y valor de producción
agrupado = chiapas.groupby(['Anio', 'Nomcultivo']).agg({
    'Volumenproduccion': 'sum',
    'Valorproduccion': 'sum'
}).reset_index()

# Paso 2: Filtrar los cultivos más importantes (opcional)
# Puedes hacer un análisis general o enfocarte en los cultivos más importantes
cultivos_importantes = agrupado.groupby('Nomcultivo')['Volumenproduccion'].sum().nlargest(5).index
agrupado_filtrado = agrupado[agrupado['Nomcultivo'].isin(cultivos_importantes)]

# Paso 3: Graficar la tendencia de volumen de producción a lo largo del tiempo
plt.figure(figsize=(14, 7))
for cultivo in cultivos_importantes:
    cultivo_data = agrupado_filtrado[agrupado_filtrado['Nomcultivo'] == cultivo]
    plt.plot(cultivo_data['Anio'], cultivo_data['Volumenproduccion'], label=cultivo)

plt.title('Tendencia del Volumen de Producción por Cultivo (2003-2023)')
plt.xlabel('Año')
plt.ylabel('Volumen de Producción (ton)')
plt.legend()
plt.grid(True)
plt.show()

# Paso 4: Graficar la tendencia de valor de producción a lo largo del tiempo
plt.figure(figsize=(14, 7))
for cultivo in cultivos_importantes:
    cultivo_data = agrupado_filtrado[agrupado_filtrado['Nomcultivo'] == cultivo]
    plt.plot(cultivo_data['Anio'], cultivo_data['Valorproduccion'], label=cultivo)

plt.title('Tendencia del Valor de Producción por Cultivo (2003-2023)')
plt.xlabel('Año')
plt.ylabel('Valor de Producción (pesos)')
plt.legend()
plt.grid(True)
plt.show()
```
