```python
# Filtrar el DataFrame para incluir solo los principales cultivos
chiapas_filtrado = chiapas[chiapas['Nomcultivo'].isin(['Pastos y praderas', 'Palma africana o de aceite', 'Palma camedor'])]

# Agrupar por cultivo y año, calculando el precio promedio
precio_por_cultivo_ano = chiapas_filtrado.groupby(['Nomcultivo', 'Anio']).agg({
    'Precio': 'mean'
}).reset_index()
```

```python
# Graficar la tendencia de los precios a lo largo del tiempo para los cultivos seleccionados
plt.figure(figsize=(14, 7))
for cultivo in ['Pastos y praderas', 'Palma africana o de aceite', 'Palma camedor']:
    data_cultivo = precio_por_cultivo_ano[precio_por_cultivo_ano['Nomcultivo'] == cultivo]
    plt.plot(data_cultivo['Anio'], data_cultivo['Precio'], marker='o', label=cultivo)

plt.title('Evolución de Precios de los Principales Cultivos (2003-2023)')
plt.xlabel('Año')
plt.ylabel('Precio Promedio (Pesos por Tonelada)')
plt.legend(title="Cultivo")
plt.grid(True)
plt.show()
```

```python
# Crear un índice para las barras
indices = range(len(df))

# Crear el gráfico de barras con los índices como posiciones
plt.figure(figsize=(8, 5))
bars = plt.bar(indices, df['porcentaje de pobreza'], color='skyblue')

# Añadir títulos y etiquetas
plt.title('Porcentaje de Pobreza por Año')
plt.xlabel('Año')
plt.ylabel('Porcentaje de Pobreza (%)')

# Asignar etiquetas manualmente para que solo los años aparezcan en el eje X
plt.xticks(indices, df['Año'])

# Añadir etiquetas de los valores sobre cada barra
for i, bar in enumerate(bars):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
             f'{df["porcentaje de pobreza"][i]:.2f}%', ha='center', va='bottom')

# Mostrar el gráfico
plt.show()
```
